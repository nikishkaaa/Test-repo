```python
def extract_hour(date_string):
    datetime_object = datetime.datetime.strptime(date_string, '%Y-%m-%d %H:%M:%S')
    return datetime_object.hour

```

```python
df['date_trunc'] = df['date_trunc'].astype(str)
```

```python
df['time'] = df['date_trunc'].apply(extract_hour)
```

```python
df.head()
```

```python
sns.countplot(df, x=df['time'], palette="ch:.25")
```
